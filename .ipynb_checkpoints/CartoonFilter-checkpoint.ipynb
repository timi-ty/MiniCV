```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Convenience function to display image
def showImage(image, isColor = True):
    plt.figure(figsize=(8, 6))
    if(isColor):
        plt.imshow(image)
    else:
        plt.imshow(image, cmap='gray')
```

```python
def applyKernel(image, kernel):
    
    output = cv2.filter2D(image,-1,kernel)
    
    return output
```

```python
def applyMask(image, mask):
    return cv2.bitwise_and(image, mask)
```

```python
#Read image in
image_path = "./man.jpg"
raw_image = cv2.imread(image_path)
raw_image = cv2.cvtColor(raw_image, cv2.COLOR_BGR2RGB)
showImage(raw_image)
```

<strong> Cartoonize Image <strong>

In order to cartoonize the image:
1. Get edges from image
2. Greatly limit the color ranges
3. Convert the whole image to grayscale
4. Apply distinct edges back to image


```python
#This function takes in an image and genrates an edge binary mask
def edgeMask(image):
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Smoothen the image but preserve edges
    gray_filtered = cv2.bilateralFilter(gray, 7, 50, 50)

    # Apply the canny filter with medium threshold(30, 60) to extract some edges from the image
    edges = cv2.Canny(gray_filtered, 30, 60)
    
    #Create kernel and dilate the image to smoothen out the edge lines
    kernel2 = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
    edges = cv2.dilate(edges, kernel2, iterations = 3)
    
    #Invert the mask from black-white to white-black to enbale application
    ret,edges = cv2.threshold(edges, 200, 255, cv2.THRESH_BINARY_INV)
    
    return edges
```

```python
#This function simplifies the color palette by only allowing multiples of the specified factor
def snapMultipleFilter(image, factor):
    #Divide all alements of the array by the factor
    temp_array = image * (1/factor)
    
    #Force all the contents of the array to integers
    temp_array  = temp_array.astype(np.uint8)
    
    #Multiply back to get approximated but exact mutliples of the factor passed in
    image = temp_array * factor
        
    return image
```

```python
#This function brings the filters together
def cartoonizeImage(image):
    #Get the edge mask from the raw image
    mask = edgeMask(image)
    
    #Simplify the color palette of the image
    image = snapMultipleFilter(image, 90)
    
    #Convert the image to grayscale
    image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    
    #Apply the edge mask gotten above
    image = applyMask(image, mask)
    
    return image
```

<strong> Driver Code <strong>

```python
cartoon_image = cartoonizeImage(raw_image)
showImage(cartoon_image, isColor = False)
```
